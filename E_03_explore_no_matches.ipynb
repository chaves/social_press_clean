## Explore no matches

```python
import pandas as pd
pd.options.display.max_colwidth = 250
```

```python
INPUT_FILE_GZIP = './_sources_final/pickles/pression_sociale_rounds_1_2_3_corrected_search_str_opt_clean_text_dup100_clean_matches.pickle.gzip'
```

```python
df = pd.read_pickle(INPUT_FILE_GZIP, compression='gzip')
```

#### Check no matches

```python
no_matches = df[(df['str_in_text'] == False) & (df['str_in_user'] == False)]
len(no_matches)
```

```python
no_matches = df[(df['str_in_text'] == False) & (df['str_in_user'] == False) & (df['str_in_conv'] == False)]
len(no_matches)
```

```python
no_matches.head(2)
```

```python
### No matches counts
```

```python
events_no_matches_counts = no_matches.string_date.value_counts().to_frame().reset_index()
```

```python
events_no_matches_counts.columns = ['string_date', 'nbTweets']
events_no_matches_counts.head(2)
```

```python
events_counts = pd.read_excel('./_sources_final/controls/01_searchStrings_counts.xls')
```

```python
events_counts.columns = ['string_date', 'nbTweets']
events_counts.head(2)
```

```python
def get_percentage(event_str, total):
    try: 
        return int(events_no_matches_counts[events_no_matches_counts.string_date == event_str].nbTweets) / int(total) *100
    except:
        return 0
```

```python
events_counts['percent_no_matches'] = events_counts.apply(lambda row: get_percentage(row['string_date'],row['nbTweets']), axis=1)
```

```python
events_counts.sort_values('percent_no_matches', ascending=False).head(50)
```
